## 1. Veri Yapısının İncelenmesi

Bu bölümde her veri setinin kolon yapısı, veri tipi dağılımı ve temel istatistikleri incelenmiştir.

```python
def explore_dataframe(df, name):
    print(f"\n{'='*60}")
    print(f"{name} - Genel Bilgi")
    print(f"{'='*60}")

    print("\nİlk 5 Satır:")
    display(df.head())

    print("\nInfo:")
    df.info()

    print("\nSayısal Özet:")
    display(df.describe())

    print("\nKategorik Dağılım:")
    display(df.describe(include='object'))

# Uygulama
explore_dataframe(df_tahsilat, "Tahsilat")
explore_dataframe(df_tahsilat_1, "Tahsilat 1")
explore_dataframe(df_tahakkuk, "Hamamözü")
explore_dataframe(df_tahakkuk_1, "Gümüşhacıköy")
explore_dataframe(df_tahakkuk_2, "Göynücek")
```

## 2. İlçe Bazında Benzersiz Müşteri Sayısı

Tahakkuk verileri kullanılarak her ilçedeki benzersiz müşteri sayısı hesaplanmıştır.

```python
customer_counts = {
    "Hamamözü": df_tahakkuk['sozlesme_hesap_no'].nunique(),
    "Gümüşhacıköy": df_tahakkuk_1['sozlesme_hesap_no'].nunique(),
    "Göynücek": df_tahakkuk_2['sozlesme_hesap_no'].nunique()
}

customer_counts
```

```python
sum(customer_counts.values())
```

### Bulgular

- Gümüşhacıköy en yüksek müşteri sayısına sahiptir.
- Hamamözü en düşük müşteri sayısına sahiptir.
- Toplam müşteri sayısı yaklaşık 28 bin civarındadır.

## 3. Tahakkuk Verilerinin Birleştirilmesi

Tüm ilçelere ait tüketim verileri birleştirilerek toplam kayıt sayısı doğrulanmıştır.

```python
df_all = pd.concat([df_tahakkuk, df_tahakkuk_1, df_tahakkuk_2], ignore_index=True)

print("Toplam Tahakkuk Kaydı:", df_all.shape[0])
```

## 4. kWh Veri Kalitesi Analizi

Bu bölümde eksik değerler, negatif tüketimler ve aşırı uç değerler incelenmiştir.

```python
#Eksik Değer
df_all['kwh'].isna().sum()
```

```python
#Negatif Değer
negative_count = (df_all['kwh'] < 0).sum()
negative_count
```

```python
#Genel İstatistik
df_all['kwh'].describe()
```

### Bulgular

- Negatif tüketim değerleri bulunmaktadır. Bu durum sayaç düzeltmeleri veya fatura iptalleri ile ilişkili olabilir.
- Maksimum tüketim oldukça yüksektir, dağılım sağa çarpıktır.
- Ortalama ve medyan arasındaki fark, uç değerlerin ortalamayı yukarı çektiğini göstermektedir.

## 5. Hesap Sınıfına Göre Tüketim Analizi

```python
class_stats = (
    df_all
    .groupby("Hesap Sınıfı")["kwh"]
    .agg(["mean", "median", "std", "count"])
    .sort_values(by="mean", ascending=False)
)

class_stats.head(10)
```

### Bulgular

- Sanayi, arıtma tesisleri ve kamu aydınlatma sınıflarında tüketim oldukça yüksektir.
- Mesken aboneleri kayıtların büyük çoğunluğunu oluşturmaktadır ancak ortalama tüketimleri düşüktür.
- Bu dağılım ilçeler arası farkları açıklamak için önemli bir değişkendir.
